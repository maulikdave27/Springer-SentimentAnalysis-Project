Libraries used:

2. Pandas

```python
#importing neccessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading the dataset
df = pd.read_csv('result/task3_result.csv')
df.set_index('from', inplace=True) #setting employee email as index
```

```python
ranking_results = []
```

```python
#top +ve and -ve
for month in df.columns:
    #top 3 positive employees
    top_positive = df[month].sort_values(ascending=False).head(3).index.tolist()
    #top 3 negative employees
    top_negative = df[month].sort_values(ascending=True).head(3).index.tolist()
    #formatting the results
    ranking_results.append({
        'month': month,
        'top_positive': top_positive,
        'top_negative': top_negative
    })
```

```python
ranking_df = pd.DataFrame(ranking_results)
ranking_df.to_csv('result/task4_result.csv', index=False) #saving the result to a CSV file
ranking_df.head()
```
